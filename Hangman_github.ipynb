# Hangman game

```python
# Make sure to run this cell or you won't be able to use this dictionary!

hangman={0:'''
        ____________
         |



         ''',
        1:'''
        ____________
         |
         O


         ''',
        2:'''
        ____________
         |
         O
        /

        ''',
        3:'''
        ____________
         |
         O
        / \\

        ''',
        4:'''
        ____________
         |
         O
        / \\
         |
         ''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}
```

```python
# your code here

import random
from random import randrange
list_of_words = [ "FOOTBALL",
    "BASKETBALL",
    "BASEBALL",
    "SOCCER",
    "TENNIS",
    "CRICKET",
    "HOCKEY",
    "VOLLEYBALL",
    "SWIMMING",
    "ATHLETICS",
    "TABLE TENNIS",
    "BADMINTON",
    "CYCLING",
    "BOXING",
    "SKIING",
    "WRESTLING"]

word = list_of_words[randrange(len(list_of_words))]
# Initial lines for the game
line_1 = ('Welcome to Hangman Game!')
print (line_1)
print('==============================================================')

# Randomly choose a word from the list
secret_word = random.choice(list_of_words)
print("The hidden word is a", len(secret_word), "letter sports name")

# Variables to track progress
guessed_letters = ['_' for _ in range(len(secret_word))]
attempts_left = 5
correct_guesses = 0

# Main game loop
while attempts_left > 0 and correct_guesses < len(secret_word):
    print(f"\nGuess the word: {' '.join(guessed_letters)}")
    char = input('Type a letter to find the hidden word: ').upper()

    if char in secret_word:
        if char in guessed_letters:
            print(f'You already guessed the letter {char}.')
        else:
            print(f'Correct! There is a letter {char} in the secret word!')
            # Reveal the guessed letter in the word
            for index, letter in enumerate(secret_word):
                if letter == char:
                    guessed_letters[index] = char
                    correct_guesses += 1
    else:
        attempts_left -= 1
        print(f'WRONG! Number of mistakes left: {attempts_left}')
        print(hangman[6 - attempts_left])
print('==============================================================')
# Check if the player won or lost
if correct_guesses == len(secret_word):
    print(f'\nCongratulations! You guessed the word: {secret_word}')
else:
    print(f'\nGame over! Better luck next time! The correct word was: {secret_word}')
```
